# Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
def check_number(n):
    if n>0:
        return "Positive Number"
    elif n<0:
        return "Negative Number"
    else:
        return "Its a Zero"
    
print(check_number(5))
print(check_number(-5))
print(check_number(0))
```

# Write a Python Program to Check if a Number is Odd or Even?

```python
def check_evenodd(n):
    if n%2==0:
        return "Even Number"
    else:
        return "Odd Number"
    
n = int(input("Enter the number you want to check "))
print(check_evenodd(n))
```

# Write a Python Program to Check Leap Year?

### Leap Years are any year that can be exactly divided by 4 (such as 2016, 2020, 2024, etc) not except if it can be exactly divided by 100, then it isn't (such as 2100, 2200, etc) yes except if it can be exactly divided by 400, then it is (such as 2000, 2400).

```python
def check_leap(year):
    if year%4==0:
        if year%100 != 0:
            return "Yes its a Leap Year"
        else:
            if year%400 == 0:
                return "Yes its a Leap Year"
            else:
                return "No it is not a leap year"        
    else:
        return "No it is not a leap year"
    
print(check_leap(2024))
print(check_leap(2100))
print(check_leap(2000))
print(check_leap(2026))
```

# Write a Python Program to Check Prime Number?

```python
def check_prime(num):
    for i in range(2,num):
        if num%i==0:
            return True
    return False

if not check_prime(int(input("Enter the number you want to check"))):
    print("Yes the number is Prime")
else:
    print("The number is not Prime")
```

# Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
def check_prime(num):
    for i in range(2,num):
        if num%i==0:
            return True
    return False

for i in range(1,10000):
    if not check_prime(i):
        print(f"The number {i} is Prime")
    else:
        print(f"The number {i} is not Prime")
        
```
